```python
# Constants
CEMENT_WEIGHT_MIN = 24.9
CEMENT_WEIGHT_MAX = 25.1
GRAVEL_SAND_WEIGHT_MIN = 49.9
GRAVEL_SAND_WEIGHT_MAX = 50.1
CEMENT_PRICE = 3
GRAVEL_SAND_PRICE = 2
SPECIAL_PACK_PRICE = 10

# Global variables to store order details
num_sacks_cement = 0
num_sacks_gravel = 0
num_sacks_sand = 0

# Task 1 - Check the contents and weight of a single sack
def check_sack():
    contents = input("Enter contents (C for cement, G for gravel, S for sand): ").upper()
    try:
        weight = float(input("Enter weight of the sack in kilograms: "))
    except ValueError:
        print("Invalid weight input. Please enter a numeric value.")
        return False

    if contents == "C" and CEMENT_WEIGHT_MIN <= weight <= CEMENT_WEIGHT_MAX:
        print("Sack accepted: Cement")
        return True
    elif contents in ["G", "S"] and GRAVEL_SAND_WEIGHT_MIN <= weight <= GRAVEL_SAND_WEIGHT_MAX:
        print(f"Sack accepted: {contents}")
        return True
    else:
        print("Sack rejected. Reasons:")
        if contents not in ["C", "G", "S"]:
            print("- Invalid contents")
        if contents == "C" and not CEMENT_WEIGHT_MIN <= weight <= CEMENT_WEIGHT_MAX:
            print("- Invalid weight for cement")
        elif contents in ["G", "S"] and not GRAVEL_SAND_WEIGHT_MIN <= weight <= GRAVEL_SAND_WEIGHT_MAX:
            print("- Invalid weight for gravel/sand")
        return False

# Task 2 - Check a customer’s order for delivery
def check_order():
    global num_sacks_cement, num_sacks_gravel, num_sacks_sand
    total_weight = 0
    rejected_sacks = 0

    try:
        num_sacks_cement = int(input("Enter the number of sacks of cement: "))
        num_sacks_gravel = int(input("Enter the number of sacks of gravel: "))
        num_sacks_sand = int(input("Enter the number of sacks of sand: "))
    except ValueError:
        print("Invalid input. Please enter a numeric value for the number of sacks.")
        return

    for _ in range(num_sacks_cement):
        if not check_sack():
            rejected_sacks += 1
        else:
            total_weight += CEMENT_WEIGHT_MAX

    for _ in range(num_sacks_gravel):
        if not check_sack():
            rejected_sacks += 1
        else:
            total_weight += GRAVEL_SAND_WEIGHT_MAX

    for _ in range(num_sacks_sand):
        if not check_sack():
            rejected_sacks += 1
        else:
            total_weight += GRAVEL_SAND_WEIGHT_MAX

    print(f"Total weight of the order: {total_weight} kilograms")
    print(f"Number of sacks rejected from the order: {rejected_sacks}")

# Task 3 - Calculate the price for a customer’s order
def calculate_price():
    regular_price = (
        num_sacks_cement * CEMENT_PRICE +
        (num_sacks_gravel + num_sacks_sand) * GRAVEL_SAND_PRICE
    )
    num_special_packs = min(num_sacks_cement, num_sacks_gravel // 2, num_sacks_sand // 2)
    discount_price = num_special_packs * SPECIAL_PACK_PRICE

    final_price = regular_price - discount_price

    print(f"Regular price for the order: ${regular_price}")
    if discount_price > 0:
        print(f"Discount price for special packs: ${discount_price}")
        print(f"New price for the order: ${final_price}")
        print(f"Amount saved: ${discount_price}")

# Test the program
check_order()
calculate_price()
```
