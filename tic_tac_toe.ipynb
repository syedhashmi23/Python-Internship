```python
def board_print(board):
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("---------")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("---------")
    print(f"{board[6]} | {board[7]} | {board[8]}")

def win_check(board):
    for i in range(0,9,3): #check horizontal
        if board[i]==board[i+1]==board[i+2] != " ":
            return True
    for i in range(0,3): #check vertical
        if board[i]==board[i+3]==board[i+6] != " ":
            return True
    if board[0]==board[4]==board[8] != " ": #check diagonals
        return True
    if board[2]==board[4]==board[6] != " ":
        return True
    return False

def tie_check(board):
    for i in range(9):
        if board[i] == " ":
            return False
    return True

board = [" "]*9
player = "X"
board_print(board)

while not win_check(board) and not tie_check(board):
    move = input(f"{player}'s turn. Choose a position from 1 to 9: ")
    if board[int(move)-1] == " ":
        board[int(move)-1] = player
        board_print(board)
        player = "O" if player == "X" else "X"
    else:
        print("Invalid move.")

if win_check(board):
    print(f"{player} wins!")
else:
    print("It's a tie!")

```
