```python
items = {
    "Case": {"A1": ("Compact", 75.00), "A2": ("Tower", 150.00)},
    "RAM": {"B1": ("8 GB", 79.99), "B2": ("16 GB", 149.99), "B3": ("32 GB", 299.99)},
    "Main Hard Disk Drive": {
        "C1": ("1 TB HDD", 49.99),
        "C2": ("2 TB HDD", 89.99),
        "C3": ("4 TB HDD", 129.99),
    },
    "Solid State Drive": {"D1": ("240 GB SSD", 59.99), "D2": ("480 GB SSD", 119.99)},
    "Second Hard Disk Drive": {
        "E1": ("1 TB HDD", 49.99),
        "E2": ("2 TB HDD", 89.99),
        "E3": ("4 TB HDD", 129.99),
    },
    "Optical Drive": {"F1": ("DVD/Blu-Ray Player", 50.00), "F2": ("DVD/Blu-Ray Re-writer", 100.00)},
    "Operating System": {"G1": ("Standard Version", 100.00), "G2": ("Professional Version", 175.00)},
}
chosen_items = {"Case": None, "RAM": None, "Main Hard Disk Drive": None}
additional_items = []
total_price = 0

def select_item(category):
    print(f"Available {category} options:")
    for code, (description, price) in items[category].items():
        print(f"{code}: {description} - ${price}")

    while True:
        user_choice = input(f"Select {category} item code: ").strip().upper()
        if user_choice in items[category]:
            chosen_items[category] = items[category][user_choice]
            globals()["total_price"] += chosen_items[category][1]
            break
        else:
            print("Invalid. Please try again.")

def select_additional_items():
    print("Do you want to purchase additional items? (yes/no)")
    choice = input().strip().lower()
    while choice == "yes":
        print("Available categories for additional items:")
        for key in items:
            if key not in chosen_items:
                print(key)

        category_choice = input("Select a category: ").strip()

        if category_choice in items and category_choice not in chosen_items:
            select_item(category_choice)
            additional_items.append(chosen_items[category_choice])
            print("Item added.")
        else:
            print("Invalid category. Please try again.")

        print("Do you want to purchase more additional items? (yes/no)")
        choice = input().strip().lower()

def calculate_discount():
    additional_items_count = len(additional_items)
    discount = 0
    if additional_items_count == 1:
        discount = 0.05
    elif additional_items_count >= 2:
        discount = 0.1

    discount_amount = discount * total_price
    globals()["total_price"] -= discount_amount

    print(f"Amount saved due to discount: ${discount_amount}")
    print(f"Total price after discount: ${total_price}")

print("Welcome to the Computer Shop!")
select_item("Case")
select_item("RAM")
select_item("Main Hard Disk Drive")

print("Chosen items:")
for category, item in chosen_items.items():
    print(f"{category}: {item[0]} - ${item[1]}")

total_price += sum(item[1] for item in chosen_items.values())
print(f"Initial total price: ${total_price}")

select_additional_items()

print("Additional items:")
for item in additional_items:
    print(f"{item[0]} - ${item[1]}")

total_price += sum(item[1] for item in additional_items)
print(f"Updated total price: ${total_price}")

calculate_discount()

```
